```python
print(df['Gender'].value_counts())
print(df['CampaignChannel'].value_counts())
print(df['CampaignType'].value_counts())
```

```python
def age_cat(df):
  if  df['Age']<12:
    return 'Kid'
  elif 13<df['Age']<=18:
    return 'Teen'
  elif 19<df['Age']<=30:
    return 'Young Adult'
  elif 31<df['Age']<=50:
    return 'Adult'
  else:
    return 'Old'
```

```python
df['Age_category'] = df.apply(age_cat,axis=1)
```

```python
df
```

```python
df.groupby(['Age_category','CampaignChannel'])['PreviousPurchases'].mean().reset_index().sort_values(by=['Age_category','PreviousPurchases'],ascending=False)
```

Conclusion
- If age category is - Effective Medium
- young adult - Social Media
- Teen - Social Media
- Adult - PPC (Pay Per Click)
- Old - Email

```python
gender_campaign_con = df.groupby(['Gender','CampaignChannel','CampaignType'])['Conversion'].agg(['sum','size']).reset_index()
```

```python
gender_campaign_con['convert_pr'] = gender_campaign_con['sum']/gender_campaign_con['size']*100
```

```python
gender_campaign_con.sort_values(by=['Gender','CampaignType'],ascending=False)
```

CONCLUSIONS
For all gender and for all campaign type , all type of campaign channels are being used.

- Male (Awareness)-Referral
- Female(Awareness)- Referral
- Male (Consideration)-PPC
- Female(Consideration)- Social Media
- Male (Conversion)-REFERRAL
- Female(Conversion)- SEO
- Male (Retention)-Email
- Female(Retention)- Referral
